# class imbalance means that the target variable has a disproportionate distribution of classes. In this case, the ' education' column has a significant imbalance,
#  with the majority of entries being 'Graduate' and very few entries for 'Not Graduate'. This can lead to biased model performance, 
# as the model may learn to predict the majority class more accurately while neglecting the minority class. To address

```python
# there are not  any class imbalance in the target variable so we can proceed with the model building
cat_cols = df.select_dtypes(include='str').columns
le = LabelEncoder()
for i in cat_cols:
    df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
X = df.drop(' loan_status', axis=1)
Y = df[' loan_status']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=56)
dt= DecisionTreeClassifier(max_depth=4, random_state=56)
dt.fit(X_train, Y_train)
```

```python
pred_Y  = dt.predict(X_test) 
print("Classification Report:",classification_report(Y_test, pred_Y))
print("Confusion Matrix:",confusion_matrix(Y_test, pred_Y))
print("Accuracy Score:",accuracy_score(Y_test, pred_Y))
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(
    dt,
    feature_names=X.columns,
    class_names=['Approved', 'Not Approved'],
    filled=True
)
plt.show()
```
